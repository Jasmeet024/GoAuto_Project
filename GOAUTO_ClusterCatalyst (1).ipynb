```python
# identifying the outliers using IQR
# Selecting numerical columns
numerical_columns = df1.select_dtypes(include=['float64', 'int64']).columns

# Calculating the first (Q1) and third (Q3) quartiles for each column
Q1 = df1[numerical_columns].quantile(0.25)
Q3 = df1[numerical_columns].quantile(0.75)

# Calculating the Interquartile Range (IQR)
IQR = Q3 - Q1

# Defining lower and upper bounds for detecting outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Finding outliers by comparing values against bounds
outliers = (df1[numerical_columns] < lower_bound) | (df1[numerical_columns] > upper_bound)

# Displaying the number of outliers per column
outliers_count = outliers.sum()
print("Number of outliers per column:")
print(outliers_count)
```

again we are not using below graph , its just to check the outliers


```python
# visualizing the outliers with boxplots
import seaborn as sns
import matplotlib.pyplot as plt

# Creating boxplots for each numerical column to visualize the outliers
for col in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df1[col])
    plt.title(f'Boxplot of {col}, Before ')
    plt.show()
```

```python
# marking outliers on the scatter plots
for col in numerical_columns:
    plt.figure(figsize=(8, 6))

    # Marking the outliers
    df1['outlier'] = outliers[col]

    sns.scatterplot(data=df1, x=df1.index, y=col, hue='outlier', palette={True: 'red', False: 'blue'})
    plt.title(f'Scatter plot of {col} with Outliers')
    plt.show()
```

```python
# Cap outliers at the upper and lower bounds
df2 = df1.copy()
for col in numerical_columns:
    df2[col] = np.where(df2[col] > upper_bound[col], upper_bound[col],
                               np.where(df2[col] < lower_bound[col], lower_bound[col], df2[col]))

print("Outliers have been capped.")
```

```python
top_makes = df2['make'].value_counts().index[:6]
df_filtered = df2[df2['make'].isin(top_makes)]

# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 8))  # 2 rows, 3 columns
axes = axes.flatten()  # Flatten to iterate easily

# Plot histogram for each top make
for i, make in enumerate(top_makes):
    ax = axes[i]
    sns.histplot(df_filtered[df_filtered['make'] == make]['price'], bins=20, kde=True, ax=ax, color=sns.color_palette("tab10")[i])
    ax.set_title(f"Price Distribution - {make}")
    ax.set_xlabel("Price ($)")
    ax.set_ylabel("Count")

# Adjust layout
plt.tight_layout()
plt.show()
```
